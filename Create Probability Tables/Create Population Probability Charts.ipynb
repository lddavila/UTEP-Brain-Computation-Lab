```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def createPopulationProbabilityChart(givenClusterNames,givenProbabilities,plotTitle,givenFileName):
    plt.bar(givenClusterNames,givenProbabilities)
    plt.title(plotTitle)
    plt.xlabel('Clusters')
    plt.ylabel('Probabilities')
    plt.ylim(0,1)
    for i in range(len(givenClusterNames)):
        plt.text(i,givenProbabilities[i],"{0:.2f}".format(givenProbabilities[i]),ha='center')
    plt.savefig(givenFileName)
    plt.show
```

```python
lg_etoh = pd.read_csv("lg_etoh_cluster_probabilities.csv")
clusterNames = lg_etoh.Cluster_Names.values.tolist()
lg_etoh_probs = lg_etoh.Probabilities.values.tolist()
```

```python
createPopulationProbabilityChart(clusterNames,lg_etoh_probs,
                                 "lg_etoh PopulationProbabilities",
                                 "lg_etoh_probability_graph.pdf")
```

```python
lg_boost = pd.read_csv('lg_boost_cluster_probabilities.csv')
lg_boost_probs = lg_boost.Probabilities.values.tolist()
createPopulationProbabilityChart(clusterNames,lg_boost_probs,
                                 "LG Boost Probabilities",
                                 "LG Boost Probabilities Graph.pdf")
```

```python
oxy = pd.read_csv('oxy_cluster_probabilities.csv')
oxy_probs = oxy.Probabilities.values.tolist()
createPopulationProbabilityChart(clusterNames,oxy_probs,
                                 "Oxy Probabilities",
                                 "Oxy Probabilities Graph.pdf")
```
