```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
def showMaleVsFemale(clusterNames,MaleProbs,FemaleProbs,experiment):
    figure(figsize=(20,10),dpi=100)
    plt.subplot(1,2,1)
    plt.bar(clusterNames,MaleProbs)
    plt.title(experiment+' Male Probabilities')
    plt.xlabel('Clusters')
    plt.ylabel('Probabilities')
    plt.ylim(0,1)
    for i in range(len(clusterNames)):
            plt.text(i,MaleProbs[i],"{0:.2f}".format(MaleProbs[i]),ha='center')
    #plt.savefig('Baseline Gender Probabilities Bar Chart.pdf')  
    #plt.show()
    plt.subplot(1,2,2)
    plt.bar(clusterNames,FemaleProbs)
    plt.title(experiment+' Female Probabilities')
    plt.xlabel('Clusters')
    plt.ylabel('Probabilities')
    plt.ylim(0,1)
    
    for i in range(len(clusterNames)):
            plt.text(i,FemaleProbs[i],"{0:.2f}".format(FemaleProbs[i]),ha='center')
    plt.savefig(experiment +' Gender Probabilities Bar Chart.pdf')  
    plt.show()
```

```python
baselineGender = pd.read_csv("BaselineMaleVsFemale.csv")
ghrelinGender = pd.read_csv("ghrelinMaleVsFemale.csv")
foodDepGender = pd.read_csv("foodDepMaleVsFemale.csv")
```

```python
clusterNames = baselineGender.Cluster_Name.values.tolist()
baselineMaleProbs =baselineGender.Male_Probability.values.tolist()
baselineFemaleProbs =baselineGender.Female_Probability.values.tolist()

foodDepMaleProbs =foodDepGender.Male_Probability.values.tolist()
foodDepFemaleProbs =foodDepGender.Female_Probability.values.tolist()

ghrelinMaleProbs =ghrelinGender.Male_Probability.values.tolist()
ghrelinFemaleProbs =ghrelinGender.Female_Probability.values.tolist()
```

```python
showMaleVsFemale(clusterNames,baselineMaleProbs,baselineFemaleProbs,'Baseline')
showMaleVsFemale(clusterNames,foodDepMaleProbs,foodDepFemaleProbs,'Food Deprivation')
showMaleVsFemale(clusterNames,ghrelinMaleProbs,ghrelinFemaleProbs,'Ghrelin')
```
