```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import scipy.stats as stats
```

```python
data = pd.read_csv("Probabilities Table.csv")
display(data)
```

```python
def get_probabilities(row):
    return row.rat_cluster_count/row.rat_total_appeareces
```

```python
data['Probability'] = data.apply(get_probabilities,axis=1)
display(data)
```

```python
def get_name(row):
    #print(row.rat_cluster_name.split())
    return row.rat_cluster_name.split()[0]
```

```python
data['rat_name'] = data.apply(get_name,axis=1)
display(data)
```

```python
def get_cluster(row):
    #print(row.rat_cluster_name.split())
    return row.rat_cluster_name.split()[1]
```

```python
data['cluster_name'] = data.apply(get_cluster,axis=1)
display(data)
```

```python
del data["rat_cluster_name"]
```

```python
display(data)
listOfRats = data["rat_name"].values.tolist()
ratDict = {}
for rat in listOfRats:
    try:
        ratDict[rat] = ratDict[rat] +1
    except:
        ratDict[rat] = 1
ratList = list(ratDict.keys())
print(ratList)
```

```python
ratProbabilitiesDict = {}
ratClustersDict = {}
```

```python
grouped_obj = data.groupby("rat_name")
separator_string= "\n"+"-"*50+"\n"
```

```python
for rat_name, filtered_df in grouped_obj:
    print(rat_name,end=separator_string)
    ratClustersDict[rat_name] = filtered_df["cluster_name"].values.tolist()
    ratProbabilitiesDict[rat_name] = filtered_df['Probability'].values.tolist()
    print(filtered_df[["cluster_name","Probability"]].head(3),end=separator_string)
```

```python
for number in range(len(ratList)):
    travelPixelClustersList=[]
    stoppingPointsClustersList=[]
    rotationPointsClustersList= []
    reactionTimeClustersList = []
    rewardChoiceClustersList = []
    
    travelPixelProbabilitiesList=[]
    stoppingPointsProbabilitiesList=[]
    rotationPointsProbabilitiesList= []
    reactionTimeProbabilitiesList = []
    rewardChoiceProbabilitiesList = []
    
    for i in range(len(ratProbabilitiesDict[ratList[number]])):
        if 'A' in ratClustersDict[ratList[number]][i]:
            travelPixelClustersList.append(ratClustersDict[ratList[number]][i])
            travelPixelProbabilitiesList.append(ratProbabilitiesDict[ratList[number]][i])
        elif 'D' in ratClustersDict[ratList[number]][i]:
            stoppingPointsClustersList.append(ratClustersDict[ratList[number]][i])
            stoppingPointsProbabilitiesList.append(ratProbabilitiesDict[ratList[number]][i])
        elif 'G' in ratClustersDict[ratList[number]][i]:
            rotationPointsClustersList.append(ratClustersDict[ratList[number]][i])
            rotationPointsProbabilitiesList.append(ratProbabilitiesDict[ratList[number]][i])
        elif 'J' in ratClustersDict[ratList[number]][i]:
            reactionTimeClustersList.append(ratClustersDict[ratList[number]][i])
            reactionTimeProbabilitiesList.append(ratProbabilitiesDict[ratList[number]][i])
        elif 'M' in ratClustersDict[ratList[number]][i]:
            rewardChoiceClustersList.append(ratClustersDict[ratList[number]][i])
            rewardChoiceProbabilitiesList.append(ratProbabilitiesDict[ratList[number]][i])
    
    fig,ax = plt.subplots(nrows =1, ncols = 5,figsize =(40,40))
    
    z_scores = stats.zscore(travelPixelProbabilitiesList)
    
    ax[0].set_title(ratList[number]+"'s Travel Pixel Probabilities")
    ax[0].bar(travelPixelClustersList,travelPixelProbabilitiesList)
    ax[0].tick_params(axis='x',labelrotation=90)
    ax[0].set_xlabel('Cluster Names')
    ax[0].set_ylabel('Probabilities')
    ax[0].set_ylim([0, 1])
    for i in range(len(travelPixelClustersList)):
        ax[0].text(i,travelPixelProbabilitiesList[i],z_scores[i],ha='center')


    ax[1].set_title(ratList[number]+"'s Stopping Point Probabilities")
    ax[1].bar(stoppingPointsClustersList,stoppingPointsProbabilitiesList)
    ax[1].tick_params(axis='x',labelrotation=90)
    ax[1].set_xlabel('Cluster Names')
    ax[1].set_ylabel('Probabilities')
    z_scores = stats.zscore(stoppingPointsProbabilitiesList)
    ax[1].set_ylim([0, 1])
    for i in range(len(stoppingPointsClustersList)):
        ax[1].text(i,stoppingPointsProbabilitiesList[i],z_scores[i],ha='center')

    ax[2].set_title(ratList[number]+"'s Rotation Point Probabilities")
    ax[2].bar(rotationPointsClustersList,rotationPointsProbabilitiesList)
    ax[2].tick_params(axis='x',labelrotation=90)
    ax[2].set_xlabel('Cluster Names')
    ax[2].set_ylabel('Probabilities')
    z_scores = stats.zscore(rotationPointsProbabilitiesList)
    ax[2].set_ylim([0, 1])
    for i in range(len(rotationPointsClustersList)):
        ax[2].text(i,rotationPointsProbabilitiesList[i],z_scores[i],ha='center')

    ax[3].set_title(ratList[number]+"'s Reaction Time Probabilities")
    ax[3].bar(reactionTimeClustersList,reactionTimeProbabilitiesList)
    ax[3].tick_params(axis='x',labelrotation=90)
    ax[3].set_xlabel('Cluster Names')
    ax[3].set_ylabel('Probabilities')
    ax[3].set_ylim([0, 1])
    z_scores = stats.zscore(reactionTimeProbabilitiesList)
    for i in range(len(reactionTimeClustersList)):
        ax[3].text(i,reactionTimeProbabilitiesList[i],z_scores[i],ha='center')
    
    ax[4].set_title(ratList[number]+"'s Reward Choice Probabilities")
    ax[4].bar(rewardChoiceClustersList,rewardChoiceProbabilitiesList)
    ax[4].tick_params(axis='x',labelrotation=90)
    ax[4].set_xlabel('Cluster Names')
    ax[4].set_ylabel('Probabilities')
    ax[4].set_ylim([0, 1])
    z_scores = stats.zscore(rewardChoiceProbabilitiesList)
    for i in range(len(rewardChoiceClustersList)):
        ax[4].text(i,rewardChoiceProbabilitiesList[i],z_scores[i],ha='center')


    #display(rats[k])
    #display(type(rats[k]))
    #plt.title(rats[number]+"'s Cluster Appearences")
    plt.xlabel("Cluster Names")
    plt.ylabel("Probability of Appearence")
    plt.xticks(fontsize = 14,rotation = 90)
    plt.savefig(ratList[number]+".pdf")
    plt.show
    plt.close('all')
    #number+=1
```
