```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = pd.read_csv("Rat Statistics.csv")
display(data)
```

```python
rats = data.rats.values.tolist()
display(rats)
```

```python
appearences = data.appearences.values.tolist()
#display(appearences)
```

```python
listOfClusterLists = []
for i in appearences:
    listOfClusterLists.append(i.split())
```

```python
#display(listOfClusterLists)
```

```python
#display(len(listOfClusterLists))
```

```python
rats = data.rats.values.tolist()
#display(rats)
```

```python
for number in range(len(rats)):
    scores ={}
    for i in listOfClusterLists[number]:
        try:
            scores[i]+=1
        except:
            scores[i] = 1
    clusterNames = list(scores.keys())
    clusterCount = list(scores.values())
    fig,ax = plt.subplots(nrows =1, ncols = 4,figsize =(40,40))

    one_dimensional_cluster_label =[]
    one_dimensional_cluster_value = []
    two_dimensional_cluster_label = []
    two_dimensional_cluster_value = []
    three_dimensional_cluster_label = []
    three_dimensional_cluster_value = []
    four_dimensional_cluster_label = []
    four_dimensional_cluster_value = []

    for i_index,i in enumerate(clusterNames):
        if len(i) == 2:
            one_dimensional_cluster_label.append(i)
            one_dimensional_cluster_value.append(clusterCount[i_index])
        elif len(i)==4:
            two_dimensional_cluster_label.append(i)
            two_dimensional_cluster_value.append(clusterCount[i_index])
        elif len(i) ==6:
            three_dimensional_cluster_label.append(i)
            three_dimensional_cluster_value.append(clusterCount[i_index])
        elif len(i) ==8:
            four_dimensional_cluster_label.append(i)
            four_dimensional_cluster_value.append(clusterCount[i_index])
    ax[0].set_title(rats[number]+"'s 1D Cluster Appearences")
    ax[0].bar(one_dimensional_cluster_label,one_dimensional_cluster_value)
    ax[0].tick_params(axis='x',labelrotation=90)
    ax[0].set_xlabel('Cluster Names')
    ax[0].set_ylabel('Number Of Appearences')


    ax[1].set_title(rats[number]+"'s 2D Cluster Appearences")
    ax[1].bar(two_dimensional_cluster_label,two_dimensional_cluster_value)
    ax[1].tick_params(axis='x',labelrotation=90)
    ax[1].set_xlabel('Cluster Names')
    ax[1].set_ylabel('Number Of Appearences')

    ax[2].set_title(rats[number]+"'s 3D Cluster Appearences")
    ax[2].bar(three_dimensional_cluster_label,three_dimensional_cluster_value)
    ax[2].tick_params(axis='x',labelrotation=90)
    ax[2].set_xlabel('Cluster Names')
    ax[2].set_ylabel('Number Of Appearences')

    ax[3].set_title(rats[number]+"'s 4D Cluster Appearences")
    ax[3].bar(four_dimensional_cluster_label,four_dimensional_cluster_value)
    ax[3].tick_params(axis='x',labelrotation=90)
    ax[3].set_xlabel('Cluster Names')
    ax[3].set_ylabel('Number Of Appearences')


    #display(rats[k])
    #display(type(rats[k]))
    #plt.title(rats[number]+"'s Cluster Appearences")
    plt.xlabel("Cluster Names")
    plt.ylabel("Number of Appearences in each Cluster")
    plt.xticks(fontsize = 14,rotation = 90)
    plt.savefig(rats[number]+".pdf")
    plt.show
    plt.close('all')
    #number+=1
```
