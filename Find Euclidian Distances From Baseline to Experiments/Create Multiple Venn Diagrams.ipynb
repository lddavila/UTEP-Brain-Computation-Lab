```python
%matplotlib inline
import venn
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
import pandas as pd
```

```python
def createMultipleVennDiagrams(theComparisonOne,allClustersDict,):
    compDf = allClustersDict[theComparisonOne]
    theComparisonSet = set(compDf['clusterLabels'])
    for i in allClustersDict.keys():
        theCurrentdf = allClustersDict[i]
        currentSet = set(theCurrentdf['clusterLabels'])
        venn2([theComparisonSet,currentSet],(theComparisonOne,i))
        plt.show()
```

```python
def readAllClustersIntoDictionary(allClusters,clusterFilePath,fileExtension):
    namesToClusters = {}
    for i in allClusters:
        namesToClusters[i] = pd.read_excel(clusterFilePath+ "\\" +i+ fileExtension)
    return namesToClusters
```

```python
def createAVennDiagram(leftSet,rightSet,leftSetLabel,rightSetLabel,titleOfPlot):
    setLabels = [leftSetLabel,rightSetLabel]
    plt.figure()
    ax = plt.gca()
    sets = [leftSet,rightSet]
    v = venn2(sets,set_labels=setLabels,ax=ax)
    plt.title(titleOfPlot)
    plt.show()
```

```python
lSet,rSet = baselineDict["A1"],baselineDict["D1"]
lSet,rSet =set(lSet['clusterLabels']),set(rSet["clusterLabels"])
createAVennDiagram(lSet,rSet,"Baseline Travel Pixel Max Vs Shift A1","Stopping Points Max Vs Shift D1","Overlap of A1 and D1")
```
