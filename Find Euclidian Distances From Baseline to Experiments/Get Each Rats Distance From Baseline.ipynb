```python
def get_probabilities(row):
    return row.ratClusterCount/row.ratTotalAppearences
```

```python
baselineData['Probability'] = baselineData.apply(get_probabilities,axis=1)
display(baselineData)
```

```python
experimentData['Probability'] = experimentData.apply(get_probabilities,axis=1)
display(experimentData)
```

```python
def get_name(row):
    #print(row.rat_cluster_name.split())
    return row.ratClusterName.split()[0]
```

```python
baselineData['rat_name'] = baselineData.apply(get_name,axis=1)
display(baselineData)
```

```python
experimentData['rat_name'] = experimentData.apply(get_name,axis=1)
display(experimentData)
```
